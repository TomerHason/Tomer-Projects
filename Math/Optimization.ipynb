```python
#  
import numpy as np
from sympy import symbols, hessian, Function, N

x, y = symbols('x ,y')
Fxy = symbols('f', cls=Function)

Fxy = -x**2+4*x*y-5*y**2+ln(x*y)

H = hessian(Fxy, [x, y])  
H
```
