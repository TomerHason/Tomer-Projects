```python
#     

def Skew_OR_Symmetric(matrix):
    if(matrix.transpose() == matrix).all():
        return "Symmetric"
    elif(matrix.transpose() == -matrix).all():
        return "Skew-Symmetric"
    else:
        return "False"

print(Skew_OR_Symmetric(matrix))
print("------------")
symMatrix = np.array([[0,4],[4,3]])
print(symMatrix)
print("----------")
print(Skew_OR_Symmetric(symMatrix)) 
```

```python
#  
def matrix_cofactor(matrix):
    try:
        determinant = np.linalg.det(matrix)
        if(determinant!=0):
            cofactor = None
            cofactor = np.linalg.inv(matrix).T * determinant
            return cofactor
        else:
            raise Exception("singular matrix")
    except Exception as e:
        print("could not find cofactor matrix due to",e)

matrix = np.array([[2,1,1],[0,-1,2],[0,2,-1]])
print(matrix)
print("-matrix_cofactor-")
matrix_cofactor = matrix_cofactor(matrix)
print(matrix_cofactor.astype(int))

#  
print("---adj(A)---")
adj_matrix = np.transpose(matrix_cofactor)
print(adj_matrix.astype(int))
```

```python
# ?    / 
def probability_Skew_OR_Symmetric(num_of_tries):
    count = 0
    for i in range(num_of_tries):
        if Skew_OR_Symmetric(np.random.randint(-3,3,size=(2,2))) in ["Symmetric","Skew-Symmetric"]:
            count += 1
    return count/num_of_tries

print(probability_Skew_OR_Symmetric(100000))
```
