```python
import pandas as pd
```

```python
df = pd.read_excel("Strange Form.xlsx")

df["Count_Values"] = df['Amount'].str.count('\|+') + 1

df
```

```python

array_IDs = []

for i in range(df.shape[0]):
     array_IDs.extend([df["Order ID"][i]] * df["Count_Values"][i])

array_IDs


```

```python
def rows_to_column(DataFrameName,column_name):
    import pandas as pd
    df = pd.read_excel(DataFrameName+".xlsx")

    split_Amount = df[column_name].str.split(pat='|',expand=True) # Split to columns base on | symbol 

    Transpose_Amount = split_Amount.T # Transpose dataset after the split 

    union_the_columns = pd.Series(Transpose_Amount.values.ravel('F')).dropna().reset_index() # union & dropna 

    union_the_columns = union_the_columns.rename(columns = {"index":"index",0:column_name}) # rename

    union_the_columns = union_the_columns[[column_name]] # select

    return union_the_columns

Amount_fix = rows_to_column("Strange Form","Amount")

Category_fix = rows_to_column("Strange Form","Category")

df = pd.DataFrame(array_IDs,columns=['Order ID'])

df = pd.concat([df,Category_fix,Amount_fix], axis=1)
df
```
