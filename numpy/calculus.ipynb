```python
# ייבוא ספריות
import numpy as np
import matplotlib.pyplot as plt
```

```python
# יצירת פרבולה

x = np.linspace(-100,100,100)
y = x**2

plt.plot(x,y)
```

```python
# נגזרת הפרבולה

dydx = np.gradient(y,x)

plt.plot(x,y)
plt.plot(x,dydx)
```

```python
# אינטגרל [0,2]

#         ↡↡↡--dy--↡↡↡    ↡↡↡--dx--↡↡↡
integral = np.cumsum(y) * (x[1] - x[0])

plt.plot(integral)
```

```python
# נצייר ספינר

theta = np.linspace(0,2*np.pi,40) #theta in[0,2pi]

Fidget_Spinner = 1 + 3/4 *np.sin(3*theta)

# נעביר לצורה קרטזית
x = Fidget_Spinner*np.cos(theta)

y = Fidget_Spinner*np.sin(theta)

plt.plot(x,y)
```
